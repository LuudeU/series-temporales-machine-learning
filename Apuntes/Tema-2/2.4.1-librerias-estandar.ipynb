# El módulo csv

Este módulo contiene muchas funciones útiles para leer archivos de tipo csv (comma separated values), incluso tablas de datos con otros separadores

```python
import csv #importamos el módulo
```

```python
# leyendo un archivo separado por comas con csv.reader()

with open(os.path.join(data_folder, googleFiles[0]), 'r') as fp: # handler del archivo
    print(f'leyendo el archivo "{googleFiles[0]}"...')
    reader = csv.reader(fp, delimiter=',') # creamos el objeto que leera el archivo linea por linea
    data = [] # guardaremos cada linea en una lista
    for row in reader: # recorremos cada linea
        data.append(row) # agregamos cada linea

for r in data: # imprimimos (observar que el resultado es una lista de listas)
    print(r)
```

```python
# para verificar que tenemos una lista de listas podemos imprimir algunos elementos
print(data[0])
print(data[0][1]) # debe ser 'open'
print(data[10])
print(data[10][1]) # debe ser 'open'
```

Se ha visto en el caso anterior que los nombres de las columnas se encuentran en la primera fila. Esto es así porque el archivo fue creado de ese modo. Luego, cada uno de los datos se localiza por medio de la columna (la columna 0 siempre es la fecha, la columna 1 corresponde al precio de apertura, etc). Es posible leer de otra manera los datos de modo que cada uno de ellos pueda ser reconocido por su nombre. Esto se puede hacer mediante diccionarios, donde cada fila corresponde a un diccionario con el nombre de la columna como _key_.

Para esto existe el método DictReader() que se muestra a continuación

```python
# leyendo un archivo separado por comas con csv.DictReader()

with open(os.path.join(data_folder, googleFiles[0]), 'r') as fp: # handler del archivo
    print(f'leyendo el archivo "{googleFiles[0]}"...')
    reader = csv.DictReader(fp, delimiter=',') # creamos el objeto que leera el archivo linea por linea
    data = [] # guardaremos cada linea en una lista
    for row in reader: # recorremos cada linea
        data.append(row) # agregamos cada linea

for r in data: # imprimimos (observar que el resultado es una lista de diccionarios)
    print(r)
```

```python
# veamos como acceder a algunos elementos
print(data[0]) # la primera fila es un diccionario y tiene datos
print(data[0]['Open']) # los datos se localizan a partir de las keys
```

Para guardar archivos, existen dos métodos análogos a los de lectura:

csv.Writer() y csv.DictWriter()

```python
# Ejemplo de DictWriter

destination_filename = os.path.join(os.getcwd(), 'test-csv-writer.csv') # siempre convienen los paths absolutos

with open(destination_filename, 'w') as fp: # el handler para evitar EOF y tener que cerrar
    column_names = data[0].keys() # usamos para el nombre de cada columna las keys
    writer = csv.DictWriter(fp, fieldnames=column_names, delimiter= '\t') # crearemos un archivo separado por tabuladores
    writer.writeheader() # escribimos el header del archivo
    for d in data:
        writer.writerow(d)
```

# El módulo datetime

Este módulo es sumamente importante para utilizar un tipo de dato derivado de Python. Los objetos datetime tienen definidas operaciones entre si, pueden ser convertidos fácilmente a formatos legibles por humanos entre otras ventajas

```python
import datetime # utilizamos un alias para no escribir datetime cada vez
```

```python
#para obtener el timestamp actual
ahora = datetime.datetime.now()
print(ahora, type(ahora)) # es un objeto de la clase datetime.datetime
```

```python
# se puede extraer la fecha
print(ahora.date())
# o el dia
print(ahora.day) # es un atributo por eso no lleva parentesis
# o la hora
print(ahora.time())
```
