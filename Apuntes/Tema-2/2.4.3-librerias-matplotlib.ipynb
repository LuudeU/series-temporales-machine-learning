# Vamos a hacer gráficos - Librería Matplotlib

```python
import matplotlib.pyplot as plt # un alias para no escribir mucho
import numpy as np # vamos a utilizar numpy
```

```python
# Creando un objeto Figure
figura = plt.Figure(figsize = (10,5), facecolor='lightgray', edgecolor='black')
type(figura)
```

Vamos a crear Axes dentro de figura

Toda la documentacion de axes esta aqui: https://matplotlib.org/api/axes_api.html?highlight=axes#module-matplotlib.axes

```python
axs = figura.subplots(1,2, sharey=True)
print('tipo: ', type(axs),'|| shape:',axs.shape)
print(axs)
```

Cada Axes va a tener un grafico diferente

```python
#Hago un array de numpy con numeros al azar uniformes
a = np.random.normal(2.5,2,1000)
#hago un histograma con 10 bins
histo_a, bins_a = np.histogram(a, bins=50)
#hago otro array con numeros random gaussianos
b = np.random.normal(2.5,2,1000)
histo_b, bins_b = np.histogram(b, bins=10)
c = np.sin(a*b)
histo_c, bins_c = np.histogram(c,bins=10)
```

```python
axs[0].plot(a,'ro',ms=3) # 'ro' significa pelotitas rojas, ms es el tamaño de la pelotita
axs[0].set_fc('whitesmoke')
axs[1].set_fc([0.9,0.9,0.98])
axs[1].hist(bins_a[:-1], bins_a, 
            weights=histo_a,fc =[0.7,0.01,0.01],ec='k',orientation='horizontal')
figura
```

## Existen otras maneras de crear subplots más simples ##

```python
fig = plt.figure(figsize=(10,6))
ax1 = fig.add_subplot(211)
ax1.plot((bins_a[:-1]+bins_a[1:])/2, histo_a, '--',lw=3)
ax1.hist(bins_a[:-1], bins_a, weights=histo_a,fc =[0.7,0.01,0.01],ec='k');
ax1 = fig.add_subplot(212)
ax1.plot((bins_b[:-1]+bins_b[1:])/2, histo_b, '--',lw=3)
ax1.hist(bins_b[:-1], bins_b, weights=histo_b,fc =[0.7,0.01,0.01],ec='k');
```

# Para hacer un gráfico rapidamente #

```python
type(plt.scatter((bins_a[:-1]+bins_a[1:])/2, histo_a))
```

# Para ajustar los límites y algunos objetos como los labels #

```python
fig = plt.figure(figsize=(10,4))
ax1 = fig.add_subplot(111)
ax1.set_ylabel('counts')
ax1.set_xlabel('x')
ax1.plot((bins_a[:-1]+bins_a[1:])/2, histo_a, '--',lw=3, label = 'linea')
ax1.hist(bins_a[:-1], bins_a, weights=histo_a,fc =[0.7,0.01,0.01],ec='k', label = 'histograma');
ax1.legend()
fig.legend()
```

```python
font = {'family' : 'Sans','weight' : 'normal', 'size'   : 18}
plt.rc('font', **font)
fig = plt.figure(figsize=(10,4))
ax1 = fig.add_subplot(111)
ax1.set_ylabel('counts')
ax1.set_xlabel('x')
ax1.set_xlim(-5,9)
ax1.plot((bins_a[:-1]+bins_a[1:])/2, histo_a, '--',lw=3, label = 'linea')
ax1.hist(bins_a[:-1], bins_a, weights=histo_a,fc =[0.7,0.01,0.01],ec='k', label = 'histograma');
ax1.legend()
```

```python
x = np.random.normal(0,2,size=30000)
y = np.random.normal(0,2,size=30000)
histo = np.histogram2d(x,y,bins=50)
fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(111)
ax.imshow(histo[0],origin='lower',extent = (x.min(), x.max() , y.min(), y.max()))
```

# Matplotlib no tiene gráficos 3D, pero tiene un constructor #

```python
from mpl_toolkits.mplot3d import Axes3D
x =np.random.normal(0,1,1000)
y =np.random.normal(0,1,1000)
z =np.random.normal(0,1,1000)
```

```python
fig = plt.figure(figsize=(8,8))
ax1 = fig.add_subplot(111, projection='3d')
ax1.set_xlim(-4,4)
ax1.set_ylim(-4,4)
ax1.set_zlim(-4,4)
ax1.scatter(x, y, z, marker='s', s=50,)
```
